```python
# Tic Tac Toe project 
# Module file if broken in
from IPython.display import clear_output


def display_board(t):
    print(f" {t[0]} | {t[1]} | {t[2]} ")
    print(f"---+---+---")
    print(f" {t[3]} | {t[4]} | {t[5]} ")
    print(f"---+---+---")
    print(f" {t[6]} | {t[7]} | {t[8]} ")


def reset_board(t):
    t = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    return t


def player_token(tokens):
    player1_token = ''
    player2_token = ''
    token = ''
    while token not in tokens:
        token = input("Player 1: Please choose a token from 'x' or 'o': ").lower()
        if token not in tokens:
            print("Please choose only 'x' or 'o'.")
            
    player1_token = token
    
    if player1_token == 'x':
        player2_token = 'o'
    else:
        player2_token = 'x'
        
    print(f"Player 1 token is '{player1_token}' and Player 2 token is '{player2_token}'. Good luck!")
    return player1_token, player2_token


def player_position():
    choice = ''
    within_range = range(1, 10)
    while not choice.isdigit() or not within_range:
        choice = input("Please choose a number between 1 and 9: ")
        
        if choice.isdigit():
            if int(choice) not in range(1, 10):
                print("Number out of range. Choose 1-9.")
                within_range = False
            else:
                within_range = True
        else:
            print("Please choose numerals only and not spell.")
            
    clear_output()
    
    return int(choice)


def game_status(token, t):
    status = ''
    display_board(t)
    
    game_dict = {'r1': [t[0], t[1], t[2]], 'r2': [t[3], t[4], t[5]], 'r3': [t[6], t[7], t[8]],
                 'c1': [t[0], t[3], t[6]], 'c2': [t[1], t[4], t[7]], 'c3': [t[2], t[5], t[8]],
                 'd1': [t[0], t[4], t[8]], 'd2': [t[2], t[4], t[6]]}
    
    for v in game_dict.values():
        if all(element == token and element != ' ' for element in v):
            status = "Win"
            print(f"\n'{token}' Wins!")
            break
    else:
        if ' ' not in t:
            status = "Draw"
            print("\nDraw")
        else:
            return False


def game_play(player1_token, player2_token, t):
    turn = 'player1'
    turn_count = 0
    
    while turn_count < 9:
        player_pos = player_position()
        if turn == 'player1':
            t[player_pos - 1] = player1_token
            status = game_status(player1_token, t)
            if status == None:
                break
            turn = 'player2'
        elif turn == 'player2':
            t[player_pos - 1] = player2_token
            status = game_status(player2_token, t)
            if status == None:
                break
            turn = 'player1'
        
        turn_count += 1
        

def gameon_choice():
    choice = 'wrong'
    
    while choice not in ['y', 'n']:
        choice = input("Would you like to play again (Y or N)? ").lower()
        
        if choice not in ['y', 'n']:
            clear_output()
            print("Sorry, I didn't understand. Please make sure to choose, Y or N.")
            
    if choice == 'y':
        return True
    else:
        return False
    
# Main file, if broken down to modules
# from tttmodules import *

def main():
    """
    A simple game of Tic Tac Toe
    """
    
    tokens = ['x', 'o']
    t = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    gameon = True
    
    while gameon:
        clear_output()
        t = reset_board(t)
        print("Let's play Tic-Tac-Toe!")
        display_board(t)
        print("First player 1 will choose a token, player 2 gets the other token.")
        player1_token, player2_token = player_token(tokens)
        game_play(player1_token, player2_token, t)
        gameon = gameon_choice()
        
        
if __name__ == "__main__":
    main()
```
